```python
def min_edit_distance(s, t):
    m = len(s)
    n = len(t)

    dp = []
    for i in range(m + 1):
        dp.append([0] * (n + 1))

    for i in range(m + 1):
        dp[i][0] = i
    for j in range(n + 1):
        dp[0][j] = j

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            cost = 0 if s[i - 1] == t[j - 1] else 1
            dp[i][j] = min(
                dp[i - 1][j] + 1,
                dp[i][j - 1] + 1,
                dp[i - 1][j - 1] + cost
            )
    return dp[m][n]

def similarity_score(distance, len1, len2):
    max_len = max(len1, len2)
    if max_len == 0:
        return 1.0
    return 1 - distance / max_len

# Program runs immediately when you run the script
word1 = input("Enter first word: ")
word2 = input("Enter second word: ")

distance = min_edit_distance(word1, word2)
similarity = similarity_score(distance, len(word1), len(word2))

print("Minimum Edit Distance:", distance)
print("Similarity Score (0 to 1):", similarity)
```
